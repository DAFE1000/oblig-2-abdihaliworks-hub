```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import brentq

def f(x):
    return np.exp(-x/4) * np.arctan(x)

def g(x):
    return np.arctan(x) - 4/(x**2 + 1)
```

```python
# finn toppunktet
x_star = brentq(g, 1, 2)
y_star = f(x_star)

print(f"x* = {x_star:.6f}")
print(f"f(x*) = {y_star:.6f}")
```

```python
x = np.linspace(-6, 12, 500)

plt.figure(figsize=(9, 5))
plt.plot(x, f(x), label=r'$f(x) = e^{-x/4}\arctan x$')
plt.plot(x_star, y_star, 'ro', label=f'Toppunkt ({x_star:.4f}, {y_star:.4f})')
plt.axhline(0, color='k', linewidth=0.7)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(r'$f(x) = e^{-x/4}\arctan x$')
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
